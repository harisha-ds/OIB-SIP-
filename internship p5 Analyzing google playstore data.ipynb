## Analyzing Google Play Store Data 

#### Description:  
Clean, categorize, and visualize Google Play Store data to understand app market dynamics.
Gain in-depth insights into the Android app market by leveraging data analytics, visualization,
and enhanced interpretation skills.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
apps = pd.read_csv('apps.csv')
reviews = pd.read_csv('user_reviews.csv')
```

```python
apps['Installs'] = apps['Installs'].astype('int')
apps['Price'] = apps['Price'].astype('float')
```

```python
merged = pd.merge(apps, reviews, on='App', how='left')
merged
```

```python
apps.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.histplot(apps['Rating'])
plt.show()
```

```python
rating=apps['Rating'].mean()
apps['Rating'].fillna(rating, inplace=True)
```

```python
plt.figure(figsize=(4,3))
sns.histplot(apps['Size'])
plt.show()
```

```python
size=apps['Size'].median()
apps['Size'].fillna(size,inplace=True)
```

```python
apps['Current Ver'] = apps['Current Ver'].fillna('Unknown')
apps['Android Ver'] = apps['Android Ver'].fillna('Unknown')
```

```python
apps.isnull().sum()
```

```python
reviews.isnull().sum()
```

```python
reviews['Translated_Review'] = reviews['Translated_Review'].fillna('No Review')
reviews['Sentiment'] = reviews['Sentiment'].fillna('Neutral')
reviews['Sentiment_Polarity'] = reviews['Sentiment_Polarity'].fillna(0.0)
reviews['Sentiment_Subjectivity'] = reviews['Sentiment_Subjectivity'].fillna(0.0)
```

```python
reviews.isnull().sum()
```

#### Category Analysis : 

```python
# App Count by Category
top_categories = apps['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title("Top 10 App Categories")
plt.xlabel("Number of Apps")
plt.show()
```

```python
# Average Rating by Category
avg_ratings = apps.groupby('Category')['Rating'].mean().sort_values(ascending=False)
avg_ratings.head(10).plot(kind='barh', color='teal')
plt.title("Top Categories by Average Rating")
plt.show()
```

#### Metrics Analysis : 

```python
# Rating vs Size
sns.scatterplot(x='Size', y='Rating', data=apps)
plt.title("App Size vs Rating")
plt.show()
```

```python
# Free vs Paid Apps
apps['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue','orange'])
plt.title("Free vs Paid Apps")
plt.show()
```

```python
#Installs vs Price
sns.scatterplot(x='Price', y='Installs', data=apps)
plt.title("Price vs Installs")
plt.show()
```

#### Sentiment Analysis : 

```python
#### Distribution of Sentiments
reviews['Sentiment'].value_counts().plot(kind='bar', color=['green','red','gray'])
plt.title("User Sentiment Distribution")
plt.show()
```

```python
# Average Polarity per Category
sentiment_by_cat = merged.groupby('Category')['Sentiment_Polarity'].mean().sort_values(ascending=False)
sentiment_by_cat.head(10).plot(kind='barh', color='purple')
plt.title("Top Categories by Average Sentiment Polarity")
plt.show()
```

## Conclusion : 

### Data Cleaning & Preparation : 

Missing values in the apps dataset were handled by imputing numerical columns (Rating, Size_MB) 
with category-wise means and medians.  

Version columns (Current Ver, Android Ver) were filled with 'Unknown' to maintain completeness.  

In the user_reviews dataset, over 26,000 entries had no review text or sentiment. 
These were labeled as 'Unknown' to differentiate between “no data” and actual neutral sentiment.  
A separate subset containing only valid reviews was used for accurate sentiment analysis.  

### Key Insights

Categories like Game, Tools, and Entertainment dominate the Play Store in both number
and installs.  

Free apps significantly outnumber Paid apps, but paid ones often have slightly higher ratings.  

App Size shows a weak relationship with Rating, suggesting user satisfaction depends more on
functionality and quality than file size.  

Sentiment analysis revealed that apps in categories such as Health & Fitness, 
Education, and Productivity often received more positive reviews.  

### Interpretation

Filling missing sentiment data with "Unknown" ensures the dataset remains complete for 
merges and visualizations, while preventing bias in sentiment statistics.  

This approach keeps the analysis transparent and fair, separating “no review” cases 
from actual sentiment-bearing data.
