```python
# Calories by Category
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Category', y='Calories', palette='Set3')
plt.title('Calories by Food Category', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.bar(df['Category'],df['Total Fat'])
plt.show()
```

Total Fat more in Chicken&fish.   

```python
# Category-wise Average Calories
avg_cal = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=avg_cal.values, y=avg_cal.index, palette='cool')
plt.title('Average Calories per Category', fontsize=14, fontweight='bold')
plt.xlabel('Average Calories')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plt.bar(df['Category'],df['Cholesterol'])
plt.show()
```

heavy sodium content observed in chicken&fish then breakfast 

```python
# Nutritional Composition
cols_to_compare = ['Calories', 'Total Fat', 'Protein', 'Sugars', 'Sodium']
sns.pairplot(df[cols_to_compare], diag_kind='kde')
plt.suptitle('Nutritional Relationships Between Key Attributes', fontsize=14, y=1.02)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
corr = numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Nutritional Factors', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# High-Calorie & Low-Calorie Items
print("\n----- Top 5 Highest Calorie Items -----")
print(df.nlargest(5, 'Calories')[['Item','Category','Calories']])

print("\n----- Top 5 Lowest Calorie Items -----")
print(df.nsmallest(5, 'Calories')[['Item','Category','Calories']])
```

High calorie foods are from breakfast and chicken&fish category and low calorie foods are from 
Beverages.  

### Business & Health Insights : 

The menu is calorie-heavy, its taste but not necessarily balanced nutritionally.

Health-conscious customers might prefer limited options like salads, fruit sides, or zero-calorie drinks.

There’s potential to introduce healthier menu items (e.g, low-fat, high-protein alternatives).

Menu labeling such as “Under 500 Calories” could improve transparency and consumer choice.  

### Conclusion : 

Overall, this analysis highlights that while McDonald’s offers a wide variety of items, many are high in calories, fats, and sodium.  
Introducing more nutritionally balanced and low-calorie options can enhance the brand’s appeal to health-conscious consumers and align better with modern dietary awareness.
